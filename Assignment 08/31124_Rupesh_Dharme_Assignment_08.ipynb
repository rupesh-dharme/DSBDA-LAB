1.	Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to see if we can find any patterns in the data.

```python
import pandas as pd
# import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic = pd.read_csv("titanic.csv")
titanic.head()
```

```python
print(titanic.shape)
```

```python
sns.jointplot(x='Fare', y='Age', data=titanic)
```

From above figure we can infer that a fair fraction of total people onboard were in age range 10-40

```python
sns.countplot(x="Age", hue="Survived", data=titanic)
```

```python
sns.distplot(titanic["Fare"], kde=False)
```

This shows that the fare was nominal for a large fraction of people

```python
sns.countplot(x="Sex", data=titanic)
# Around 560 men and 300 women
```

```python
sns.countplot(x="Pclass", hue="Survived", data=titanic)
plt.show()
```

This infer that the people  from Passenger class 1 and 2 had higher rate of survival

```python
sns.countplot(x="Sex", hue="Survived", data=titanic)
plt.show()
```

Women survived more than men

```python
tc = titanic.corr()
sns.heatmap(tc.drop(["PassengerId", "SibSp", "Parch"]).drop(["PassengerId", "SibSp", "Parch"], axis=1), cmap="coolwarm")
plt.title('titanic.corr()')
```

We can observe here as well the maximum correlation is between Fare and Survival pointing to the fact that Pclass 1 and 2 had more survival rate
