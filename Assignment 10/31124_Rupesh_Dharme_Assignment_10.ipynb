```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = pd.read_csv("Iris.csv")
iris.head()
```

```python
iris.groupby(["Species"]).head()
```

#### 1. List down the features and their types (e.g., numeric, nominal) available in the dataset.

```python
iris.dtypes
```

1. Id: numeric
2. SepalLengthCm: numeric
3. SepalWidthCm: numeric
4. PetalLengthCm: numeric
5. PetalWidthCm: numeric
6. Species: nominal

#### 2. Create a histogram for each feature in the dataset to illustrate the feature distributions.


```python
sns.histplot(x="SepalLengthCm", hue="Species", data=iris)
plt.show()
```

```python
sns.histplot(x="SepalWidthCm", hue="Species", data=iris)
plt.show()
```

```python
sns.histplot(x="PetalLengthCm", hue="Species", data=iris)
plt.show()
```

```python
sns.histplot(x="PetalWidthCm", hue="Species", data=iris)
plt.show()
```

```python
sns.histplot(x="Species", data=iris)
plt.show()
```

#### 3. Create a box plot for each feature in the dataset.
#### 4. Compare distributions and identify outliers.

```python
sns.boxplot(x="SepalLengthCm", y="Species", data=iris)
plt.show()
```

```python
sns.boxplot(x="SepalWidthCm", y="Species", data=iris)
plt.show()
```

```python
sns.boxplot(x="PetalLengthCm", y="Species", data=iris)
plt.show()
```

```python
sns.boxplot(x="PetalWidthCm", y="Species", data=iris)
plt.show()
```

```python
iris.groupby(["Species"]).describe()["SepalLengthCm"]["mean"]
```

```python
sns.jointplot(x="PetalLengthCm", y="PetalWidthCm", data=iris, hue="Species")
plt.show()
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, hue="Species")
plt.show()
```
